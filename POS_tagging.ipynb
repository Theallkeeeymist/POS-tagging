```python
import spacy
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
doc=nlp(u"Hi! I am Sudhanshu Anand and I am learning NLP. I yesterday learnt some architecture and I hate everything")
```

```python
doc[4].pos_
```

```python
doc[4].tag_
```

```python
for word in doc:
  print(word.text,"--->", word.pos_, ", ", spacy.explain(word.pos_), ", ", word.tag_, ", ", spacy.explain(word.tag_))
```

```python
from spacy import displacy
```

```python
displacy.render(doc,style="dep",jupyter=True)
```

```python
options={
    'distance':120,
    'compact':True,
    'color':'#fff',
    'bg':'#00a65a'
}
```

```python
displacy.render(doc, style='dep', jupyter=True, options=options)
```
